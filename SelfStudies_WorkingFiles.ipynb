# Visualizing Covariance and Correlation

## 1. Scatter Plots - The Foundation
Python Code:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Create sample data
np.random.seed(42)
hours_studied = np.random.normal(5, 2, 100)  # Average 5 hours, some variation
exam_scores = 60 + 8 * hours_studied + np.random.normal(0, 10, 100)  # Base score + improvement from studying + random noise

df = pd.DataFrame({
    'hours_studied': hours_studied,
    'exam_scores': exam_scores
})

# Basic scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['hours_studied'], df['exam_scores'], alpha=0.6)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between Study Hours and Exam Scores')
plt.show()
```

## 2. Correlation Heatmaps
Life Example: Imagine you're analyzing a student's performance across different subjects:

```python
# Create multi-subject data
np.random.seed(42)
data = {
    'Math': np.random.normal(75, 15, 100),
    'Science': np.random.normal(78, 12, 100),
    'English': np.random.normal(72, 18, 100),
    'History': np.random.normal(70, 16, 100)
}

# Make subjects somewhat correlated (good students tend to do well across subjects)
for i in range(100):
    boost = np.random.normal(0, 10)  # Some students are just generally better/worse
    data['Math'][i] += boost
    data['Science'][i] += boost * 0.7  # Science somewhat related to Math
    data['English'][i] += boost * 0.3  # English less related to Math
    data['History'][i] += boost * 0.4

student_df = pd.DataFrame(data)

# Create correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = student_df.corr()
sns.heatmap(correlation_matrix, 
            annot=True,  # Show correlation values
            cmap='RdBu',  # Red-Blue color scheme
            center=0,    # Center colormap at 0
            square=True)
plt.title('Subject Performance Correlation Matrix')
plt.show()
```

## 3. Pair Plots - Multiple Relationships at Once

```python
# Pair plot to see all relationships
sns.pairplot(student_df)
plt.suptitle('All Subject Relationships', y=1.02)
plt.show()
```

## 4. Different Correlation Patterns

```python
# Create datasets showing different correlation patterns
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Perfect positive correlation
x1 = np.linspace(0, 10, 50)
y1 = 2 * x1 + np.random.normal(0, 0.1, 50)
axes[0,0].scatter(x1, y1)
axes[0,0].set_title(f'Strong Positive Correlation\nr = {np.corrcoef(x1, y1)[0,1]:.2f}')

# Perfect negative correlation  
x2 = np.linspace(0, 10, 50)
y2 = -2 * x2 + 20 + np.random.normal(0, 0.1, 50)
axes[0,1].scatter(x2, y2)
axes[0,1].set_title(f'Strong Negative Correlation\nr = {np.corrcoef(x2, y2)[0,1]:.2f}')

# No correlation
x3 = np.random.normal(5, 2, 50)
y3 = np.random.normal(5, 2, 50)
axes[1,0].scatter(x3, y3)
axes[1,0].set_title(f'No Correlation\nr = {np.corrcoef(x3, y3)[0,1]:.2f}')

# Non-linear relationship (correlation might be misleading!)
x4 = np.linspace(-3, 3, 50)
y4 = x4**2 + np.random.normal(0, 0.5, 50)
axes[1,1].scatter(x4, y4)
axes[1,1].set_title(f'Non-linear (Correlation = {np.corrcoef(x4, y4)[0,1]:.2f})')

plt.tight_layout()
plt.show()
```

# datetime Module

```python
from datetime import date, datetime, timedelta

today = date.today()
print("Today:", today)  # e.g., 2025-07-27

now = datetime.now()
print("Now:", now)

formatted = now.strftime("%A, %d %B %Y %H:%M:%S")
print("Formatted:", formatted)

parsed = datetime.strptime("15 August, 2025", "%d %B, %Y")
print("Parsed datetime:", parsed)

future = now + timedelta(days=10, hours=5)
print("Future datetime:", future)
```
